<font size = "3" >

- Read the dataset labelled features.csv.

value 2588
- Subset the data to get one dataset with weight $>$ threshold and <br>
another with weight $\le$ 2588. Use the ```.query()``` function.
- Compute the mean acceleration for each of these datasets.<br>

Note: Review how to define a gloabl variable inside ```.query()```
 

````python
# Write your answer here:
datasetb = pd.read_csv('data/features.csv')
threshold = 2588
dataabove = datasetb.query('weight > @threshold')
databelow = datasetb.query('weight <= @threshold')
mean_acceleration1 = dataabove['acceleration'].mean()
mean_acceleration2 = databelow['acceleration'].mean()
print(f"Mean acceleration for cars with weight > {threshold}: {mean_acceleration1}")
print(f"Mean acceleration for cars with weight <= {threshold}: {mean_acceleration2}")
````

(e) Plot by Category

<font size = "3">

Read the "data/wdi_urban.csv" dataset.

  This contains data from 1980 and 2020.

- **Create a list with the unique values of "year"**

- **Create a plot with a ...**
    - Histogram of "prop_urbanpopulation" for each year
    - Set the histogram option `alpha = 0.5` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title


````python
# Write your answer here:
datasete = pd.read_csv('data/wdi_urban.csv')
unique_years = pd.unique(datasete['year'])
plt.figure(figsize=(6, 6))
for year in unique_years:
    subset = datasete[datasete['year'] == year]
    plt.hist(subset['prop_urbanpopulation'], alpha=0.5, label=str(year))
plt.legend(title='Year')
plt.xlabel('Proportion of Urban Population')
plt.ylabel('Frequency')
plt.title('Proportion Of The Urban Population In Each Year')
plt.show()
````
